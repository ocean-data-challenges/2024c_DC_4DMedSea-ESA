The figure shows that the maximum mapping errors are found in intense current systems, for example in the GulfStream, Kuroshio and Agulhas regions. 

However, when considering the full scale of motion in the drifter database, the surface current maps capture up to 80% of the variability of drifter currents in the Western Boundary Currents and Antarctic Circumpolar Currents (ACC). The geostrophic signal dominates the ageostrophic signal in these regions. In regions with low ocean variability, only a few percent of the total drifter current variability is recovered in the maps, which may be associated with a larger ageostrophic signal in these regions.

## 2.4 Compute Spectral scores

```python
# Compute PSD scores
compute_psd_scores_current(ds_interp, psd_output_filename, lenght_scale=segment_lenght)
```

```python
id_sensor = np.unique(ds_interp.sensor_id)
seg_size = 100

list_seg_EWCT = list()
list_seg_NSCT = list()
list_seg_ugos_interpolated = list()
list_seg_vgos_interpolated = list()
list_seg_ugos_diff = list()
list_seg_vgos_diff = list()

for id_s in id_sensor: 
    ds_interp_id = ds_interp.where(ds_interp.sensor_id == id_s, drop=True)
    if np.size(ds_interp_id.EWCT)>seg_size :
        list_seg_EWCT.append(ds_interp_id.EWCT[:100])
        list_seg_NSCT.append(ds_interp_id.NSCT[:100])
        list_seg_ugos_interpolated.append(ds_interp_id.ugos_interpolated[:100])
        list_seg_vgos_interpolated.append(ds_interp_id.vgos_interpolated[:100])
        
        list_seg_ugos_diff.append(ds_interp_id.ugos_interpolated[:100]-ds_interp_id.EWCT[:100])
        list_seg_vgos_diff.append(ds_interp_id.vgos_interpolated[:100]-ds_interp_id.NSCT[:100])
```

```python

wn, psd_EWCT = sig.welch(np.array(list_seg_EWCT).ravel(),fs=24/6,nperseg=seg_size,scaling='density',noverlap=0)
wn, psd_NSCT = sig.welch(np.array(list_seg_NSCT).ravel(),fs=24/6,nperseg=seg_size,scaling='density',noverlap=0)
wn, psd_ugos = sig.welch(np.array(list_seg_ugos_interpolated).ravel(),fs=24/6,nperseg=seg_size,scaling='density',noverlap=0)
wn, psd_vgos = sig.welch(np.array(list_seg_vgos_interpolated).ravel(),fs=24/6,nperseg=seg_size,scaling='density',noverlap=0)
wn, psd_ugos_d = sig.welch(np.array(list_seg_ugos_diff).ravel(),fs=24/6,nperseg=seg_size,scaling='density',noverlap=0)
wn, psd_vgos_d = sig.welch(np.array(list_seg_vgos_diff).ravel(),fs=24/6,nperseg=seg_size,scaling='density',noverlap=0)
```

```python

wn, psd_EWCT = sig.welch(np.array(list_seg_EWCT).ravel(),fs=24/6,nperseg=seg_size,scaling='density',noverlap=0)
wn, psd_NSCT = sig.welch(np.array(list_seg_NSCT).ravel(),fs=24/6,nperseg=seg_size,scaling='density',noverlap=0)
wn, psd_ugos2 = sig.welch(np.array(list_seg_ugos_interpolated).ravel(),fs=24/6,nperseg=seg_size,scaling='density',noverlap=0)
wn, psd_vgos2 = sig.welch(np.array(list_seg_vgos_interpolated).ravel(),fs=24/6,nperseg=seg_size,scaling='density',noverlap=0)
wn, psd_ugos_d2 = sig.welch(np.array(list_seg_ugos_diff).ravel(),fs=24/6,nperseg=seg_size,scaling='density',noverlap=0)
wn, psd_vgos_d2 = sig.welch(np.array(list_seg_vgos_diff).ravel(),fs=24/6,nperseg=seg_size,scaling='density',noverlap=0)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.title('u')
plt.loglog(wn, psd_EWCT, label='drifters')
plt.loglog(wn, psd_ugos2, label=method_name)
plt.loglog(wn, psd_ugos_d2,':', label=method_name+' err')
plt.loglog(wn, psd_ugos, label="MIOST")
plt.loglog(wn, psd_ugos_d,':', label='MIOST'+' err')
plt.legend()

plt.subplot(122)
plt.title('v')
plt.loglog(wn, psd_NSCT, label='drifters')
plt.loglog(wn, psd_vgos2, label=method_name)
plt.loglog(wn, psd_vgos_d2,':', label=method_name+' err')
plt.loglog(wn, psd_vgos, label="MIOST")
plt.loglog(wn, psd_vgos_d,':', label='MIOST'+' err')
plt.legend()
```

```python
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.title('u')  
plt.semilogx(wn, 1-psd_ugos_d2/psd_EWCT,':', label=method_name+' err') 
plt.semilogx(wn, 1-psd_ugos_d/psd_EWCT,':', label='MIOST'+' err')
plt.legend()

plt.subplot(122)
plt.title('v')  
plt.semilogx(wn, 1-psd_vgos_d2/psd_NSCT,':', label=method_name+' err') 
plt.semilogx(wn, 1-psd_vgos_d/psd_NSCT,':', label='MIOST'+' err')
plt.legend()
```
